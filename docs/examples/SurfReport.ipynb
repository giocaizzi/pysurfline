# Visualize the surf report

The data stored in a `SpotForecasts` object can be visualized with a plot.

```python
import pysurfline
```

```python
spotId = "5842041f4e65fad6a7708cfd" # anchor point
# spotId = "5842041f4e65fad6a7708890" # pipeline

spotforecasts = pysurfline.get_spot_forecasts(spotId)
spotforecasts
```

The plot can be customized by calling the following arguments:
- `barLabels` (bool) : height label of surf

```python
pysurfline.plot_surf_report(
    spotforecasts,
    barLabels = True,
    )
```
